# Рекомендация тарифов

В распоряжении данные о поведении клиентов, которые уже перешли на эти тарифы (из проекта курса «Статистический анализ данных»). Нужно построить модель для задачи классификации, которая выберет подходящий тариф. Предобработка данных уже выполнена в первой части.

## 1. Загрузка и изучение файлов

```python
import pandas as pd
import seaborn as sns
import joblib

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('https://code.s3.yandex.net/datasets/users_behavior.csv')
df.info()
df.head()
```

Всего в таблице 3 214 объектов и 5 признаков для них. С данными все в порядке.

```python
a = sns.pairplot(df)
```

```python
print(df['is_ultra'].corr(df['calls']))
print(df['is_ultra'].corr(df['minutes']))
print(df['is_ultra'].corr(df['messages']))
print(df['is_ultra'].corr(df['mb_used']))
```

```python
df = df.drop(['minutes'], axis=1)
df.head()
```

Пользователи тарифа Ultra пользовались услугами мобильного оператора гораздно чаще, чем абоненты других тарифов.

На графиках видно, что есть сильная корреляция между количеством звонков и минутами разговора. Если два столбца так сильно коррелируют, это означает, что модели попросту нечему научииться у второго, поэтому удалим минуты.

## 2. Разделение данных на подвыборки

```python
df_train, df_2 = train_test_split(df, test_size=0.4, random_state=12345)
df_valid, df_test = train_test_split(df_2, test_size=0.5, random_state=12345)

print(df_train.shape)
print(df_valid.shape)
print(df_test.shape)
```

```python
features_train = df_train.drop(['is_ultra'], axis = 1)
target_train = df_train['is_ultra']

features_valid = df_valid.drop(['is_ultra'], axis = 1)
target_valid = df_valid['is_ultra']

features_test = df_test.drop(['is_ultra'], axis = 1)
target_test = df_test['is_ultra']
```

Так как спрятанной тестовой выборки нет, то данные разбили на 3 части: обучающую, валидационную и тестовую в соотношении 3:1:1.
Далее для каждого набора выделили обучающий набор данных и целевой признак.

<font color='violet'> </font>

## 3. Исследование моделей

```python
for i in range(1, 10, 1):
    model_1 = DecisionTreeClassifier(random_state=12345, max_depth=i)
    model_1.fit(features_train, target_train)
    predictions_t = model_1.predict(features_train)
    predictions_v = model_1.predict(features_valid)
    accuracy_t = accuracy_score(target_train, predictions_t)
    accuracy_v = accuracy_score(target_valid, predictions_v)
    print('max_depth:', i, 'accuracy_train: {:.2f}'.format(accuracy_t), 'accuracy_valid: {:.2f}'. format(accuracy_v))
```

Поэкспериментировали с обучаюший деревой и в цикле перебрали разное количество глубины деревьев. Лучший результат accuracy по набором (тренировочная 0.85 и валидной 0.78) при глубине узлов 7. Результат хороший, посмотрим другие модели.

```python
for i in range(1, 100, 10):
    for j in range(1, 30, 10):
        model_2 = (RandomForestClassifier(random_state=12345, n_estimators=i, max_depth=j)
                   .fit(features_train, target_train))
        predictions_t = model_2.predict(features_train)
        predictions_v = model_2.predict(features_valid)
        accuracy_t = accuracy_score(target_train, predictions_t)
        accuracy_v = accuracy_score(target_valid, predictions_v)
        print('n_estimators:', i,'max_depth:', j, 'accuracy_train: {:.2f}'.format(accuracy_t), 'accuracy_valid: {:.2f}'. format(accuracy_v))
```

Также перебрали разные гиперпараметры модели случайный лес и получили следующие результаты:
- самый высокий accuracy на двух выборках: 0.79 на валидной и 1 на тренировочной
- при параметре n_estimators 41 и глубине деревьев 21

```python
model_3 = LogisticRegression(random_state=12345, class_weight='balanced')
model_3.fit(features_train, target_train)
model_3.predict(features_valid)
model_3.score(features_valid, target_valid)
```

Результаты алгоритма логистической регрессии получаются не такими точными: accuracy всего 0.6, но она хотя бы выигрывает в скорости.

```python
model_best = (RandomForestClassifier(random_state=12345, n_estimators=41, max_depth=21)
               .fit(features_train, target_train))
```

Итак, рассмотрены три модели, для них подобраны гиперпараметры. Отдельно сохранили лучшую модель.

## 4. Проверка модели на тестовой выборке

```python
predictions_test = model_best.predict(features_test)
accuracy = accuracy_score(target_test, predictions_test)
print(accuracy)
```

Проверили модель на тестовом наборе данных, accuracy чуть ниже чем у валидационного набора, но все равно довольно высокий — целых 0.8!

## 5. Проверка модели на адекватность

```python
df_train['is_ultra'].value_counts(normalize=True)
```

```python
dummy_clf = DummyClassifier(strategy="constant", random_state=0, constant=[0])
dummy_clf.fit(features_train, target_train)
predictions_t = dummy_clf.predict(features_train)
predictions_v = dummy_clf.predict(features_valid)
accuracy_t = accuracy_score(target_train, predictions_t)
accuracy_v = accuracy_score(target_valid, predictions_v)
print('accuracy_train: {:.2f}'.format(accuracy_t), 'accuracy_valid: {:.2f}'. format(accuracy_v))
```

У элементарного способа предсказания accuracy оказалась довольна близка, но все же ниже чем у нашей модели. Но на нее также повлиял факт, что абонентов тарифа "Смарт" более чем в 2 раза больше.
