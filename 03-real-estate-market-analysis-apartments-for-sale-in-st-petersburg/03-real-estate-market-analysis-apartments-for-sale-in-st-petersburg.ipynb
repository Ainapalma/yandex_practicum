# Исследование объявлений о продаже квартир

Имеются данные сервиса Яндекс.Недвижимость — архива объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за несколько лет. По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

Задача — установить параметры определения рыночной стоимости объектов недвижимости. Они позволят построить автоматизированную систему, которая отследит аномалии и мошенническую деятельность.

## Шаг 1. Загрузка файла

```python
#импортируем библиотеки
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://code.s3.yandex.net/datasets/real_estate_data.csv', sep = '\t')
df.info()
df.head()
```

### Вывод

В архиве содержатся данные по 23 699 объявлениям, собранных по 22 параметрам. Из них большинство значений (всего 14) относятся к количественным переменным типа float64. В целом, данные переданы в аккуратном виде, изменений названия столбцов или редактирования не требуют. Приступим к исследованию данных.

## Шаг 2. Предобработка данных

```python
df.isna().sum(axis=0).sort_values(ascending=False)/len(df)
```

Посчитали количество пропусков, которые затем суммировали по всем столбцам и отсортировали по убыванию. В заключение, результат нормировали для оценки объема пропусков. Выявили упущенные значения более чем у половины объявлений в следующих стобцах:
- 'is_apartment', 
- 'parks_nearest', 
- 'ponds_nearest'.

### 2.1. Обработка пропусков

```python
df['balcony'] = df['balcony'].fillna(0)
df['is_apartment'] = df['is_apartment'].fillna(False)

df.isnull().sum()
```

Найденные пропуски были заполнены по следующей логике:
- Так как параметры 'Количество балконов' и 'Тип апартаменты' были пропущены, то скорее всего их значение отрицательное, потому как если бы она имелась, то в объявлении это наверняка бы указывалось, так как от это зависит общая площадь и, соответственно, цена жилья. По этой логике данные были заполнены нулям и булевым типом данных False;
- Оставшиеся данные оставили как есть для чистоты исследования.

### 2.2. Замена типа данных

```python
for column in ['airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 
               'ponds_nearest', 'balcony', 'floors_total']:
    df[column] = df[column].astype('Int64')

df['locality_name'] = df['locality_name'].astype(str)
df.info()
```

Заменили параметры с расстояниями (даны в метрах, поэтому сантиметры нам ни к чему), а также такие значения как Количество балконов и Этаж (у них не может быть дробных частей) с вещественного типа на целочисленные значения.

### 2.3. Обработка дубликатов

```python
df.duplicated().sum()
```

Проверили, что дубликатов нет.

## Шаг 3. Добавление новых признаков

### 3.1. Цена квадратного метра

```python
df['cost_per_meter'] =  df['last_price'] / df['total_area']
df.head()
```

Добавили стоимость за квадратный метр квартиры.

### 3.2. День недели, месяц и год публикации объявления

```python
df['date'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
df['weekday'] = df['date'].dt.weekday
df['month'] = df['date'].dt.month
df['year'] = df['date'].dt.year
df['weekday'].value_counts()
df['month'].value_counts()
df['year'].value_counts()
```

Форматировали дату размещения объявления и добавили столбцы по дню недели, месяцу и году. 
- Чаще всего объявления размещались в среду — всего 4 295 объявлений. 
- Больше всего объявлений размещали в 2018-2017 годах (8 519 и 8 190), а также в начале года Февраль-Март (2 640 и 2 587) и в конце Октябрь-Ноябрь (2 371 и 2 127). 
- В праздничные и околопраздничные месяца собиралась продавать недвижимость меньше всего: Декабрь (1 641), Январь (1 500) и Май (1 274). 

### 3.3. Этаж квартиры

```python
def floor_level(row):
    if row.floor == 1:
        return 'первый'
    if row.floor == row.floors_total:
        return 'последний'
    else:
        return 'другой'

df['floor_level'] = df.apply(floor_level, axis=1)
df['floor_level'].value_counts()
```

Категоризировали и добавили уровень этажа для каждой квартиры. Всего продающихся квартир на первом, как и на последнем этаже примерно по 3 тысячи, что вполне объяснимо: у высоток других этажей несоизмеримо много, ровно как и объявлений — всего 17 441.

```python
relation_floor = df.groupby('floor_level')['last_price'].median()
plt.bar(relation_floor.index, relation_floor)
plt.show()
```

Стоимость квартир на первом и последнем этажах немного ниже, чем на других.

### 3.4. Отношение площадей

```python
df['area_living_to_total'] = df.living_area / df.total_area
df['area_kitchen_to_total'] = df.kitchen_area / df.total_area

df.head()
```

Добавили соотношение жилой и общей площади, а также отношение площади кухни к общей.

### 3.5. Параметры: площадь, цена, число комнат, высота потолков

#### 3.5.1. Площадь

```python
df['total_area'].describe()
df.boxplot('total_area')
plt.show()
```

```python
area_less_200 = df.query('total_area < 200')
area_less_200.plot(x='total_area', y='last_price', style='o', alpha=0.4, grid=True, 
                    label='Зависимость цены от общей площади')
_=plt.axhline(75000000, color='r', linestyle='dashed', linewidth=1)
plt.show()
```

Для вычисления условия среза данных посмотрели числовое описание данных Общая площадь и затем отобразили диаграмму ящик с усами. Более чем у 75% объявлений площадь менее 70 кв.м, поэтому взяли с запасом квартиры менее 200 кв.м. Прослеживается очевидная зависимость: чем больше площадь квартиры, тем выше ее цена. Имеются выпадающие значения: слишком дорого стоят квартиры с метражом примерно 185 и 190 кв.м. Посмотрим поближе на квартиры стоимостью менее 75 млн. руб.

```python
areaLess200_costLess75 = area_less_200.query('last_price < 75000000')
areaLess200_costLess75.plot(x='total_area', y='last_price', style='o', alpha=0.4, grid=True, 
                    label='Зависимость цены от общей площади')
_=plt.axhline(35000000, color='r', ls='--', linewidth=1)
plt.show()
```

Большинство квартир площадью до 100 кв.м. и стоят до 20 млн.руб.

#### 3.5.2. Цена

```python
print(df['last_price'].corr(df['cost_per_meter']))
```

```python
areaLess200_costLess75.plot(x='cost_per_meter', y='last_price', style='-', alpha=0.7, grid=True, 
                    label='Зависимость цены от стоимости квадратного метра')
plt.show()
```

Здесь также явная зависимость общей цены от стоимости квартиры за квадратный метр: корреляция 0.74%

#### 3.5.3. Число комнат

```python
df['rooms'].value_counts().plot(kind='barh')
plt.show()
```

Большинство квартир с количеством комнат до 4, у 23х огромных квартир их больше 8. Для оценки зависимости будем использовать срез данных по квартирам с количеством комнат до 5.

```python
roomsLess5 = df.query('rooms <= 5')
roomsLess5 = roomsLess5.groupby('rooms')['last_price'].median()
plt.bar(roomsLess5.index, roomsLess5)
plt.show()
```

В данном случае тоже логичная связь: чем больше комнат, тем выше цена недвижимости

#### 3.5.4. Высота полотка

```python
df['ceiling_height'].value_counts()
```

```python
ceilingLower4 = areaLess200_costLess75.query('2 < ceiling_height < 4').plot(x='ceiling_height', y='last_price', 
                                                                            style='o', alpha=0.4, grid=True, 
                                                                            label='Зависимость цены от количества комнат')
plt.show()
```

На рынке больше всего предложений с высотой потолков 2.5-2.7 м, так как это самые популрные типы застроек. Корреляции умеренная (0,08%), по графику отчетливо видно, что квартиры с высотой потолков 3 метров и более стоят в разы больше.

### 4.3. Анализ факторов влияющих на стоимость квартиры

```python
df['cityCenters_nearest_km'] = (df['cityCenters_nearest'] / 1000).fillna(0).astype(int)
relation_cityCenter = df.groupby('cityCenters_nearest_km')['last_price'].median()
plt.bar(relation_cityCenter.index, relation_cityCenter)
plt.show()
```

Чем ближе к центру, тем выше стоимость квартиры. При этом нашли следующие артефакты: на расстояниях примерно 42 км и 53-54 км имеются пиковые значения стоимости. Возможно, это какие-то элитные загородные дома.

```python
top10 = df['locality_name'].value_counts()
top10.head(10)
```

```python
top10 = df['locality_name'].value_counts()
top10 = top10.head(10)

(df
 .where(df['locality_name'].isin(top10.index))
 .pivot_table(index='locality_name', values='cost_per_meter', aggfunc='mean')
 .sort_values('cost_per_meter')
 .plot(kind='bar', title='Cредняя цена квадратного метра в топ 10 населенных пунктах'))

plt.show()
```

Проанализировали стоимость квадратного метра в топ 10 населенных пунтках:
- Самая дорогое жилье, конечно, в Петербурге — чуть выше 100 тыс за кв.м
- Дешевле всего стоит квартира в Выборге: примерно за 60 тыс. за кв.м. можно приобрести там жилье
- В среднем 80 тыс стоит квадратный метрв оставшихся других населенных пунктах.

### 4.4. Анализ предложений

Выясним, какая область входит в центр: для этого создадим столбец с расстоянием до центра в километрах и округлим до целых значений. После этого посчитаем среднюю цену для каждого километра.

```python
StPetersburg = df.query('locality_name == "Санкт-Петербург" and cityCenters_nearest_km < 15')
(StPetersburg
 .pivot_table(index='cityCenters_nearest_km', values='last_price', aggfunc=['mean', 'median'])
 .plot(title='Зависимость цены от удалённости от центра'))

_=plt.axhline(7500000, color='r', ls='--')
```

Выделили радиус центр в Санкт-Петербурге — 15 километров и проанализировали стоимость в зависимости от удаленности от центра:
- Начиная примерно от 8 млн стоит недвижимость в центре
- Что интересно, есть небольшой спад в радиусе 3 км
- Далее 5-6 км от центра центра начинается спад стоимости

### 4.5. Анализ сегмента квартир в центре

Проанализируем территорию квартир в центре и изучим следующие параметры: площадь, цена, число комнат, высота потолков.

```python
StPetersburg = StPetersburg.query('cityCenters_nearest_km < 6')
StPetersburg = StPetersburg.pivot_table(index=list(StPetersburg.index),
                                        values=['total_area', 'last_price', 'rooms', 'ceiling_height', 'floor', 'cityCenters_nearest_km'])
    
pd.plotting.scatter_matrix(StPetersburg, figsize=(14, 14))
plt.suptitle('Анализ факторов, влияющих на стоимость квартир в центре')
plt.show()
```

В целом, тренды такие же, как и по всему городу:
- Высота потолков. Есть зависимости с количеством этажей в здании: в пятиээтажках и ниже потолки выше. Скорее всего, это красивые исторические дома, а не новые типовые застройки
- Расстояние до центра. В самом центре и на расстоянии 4 км от него есть высотки с 20-ю и более этажами.
- Количество этажей. Хотя, конечно, по общему объему преобладает количество невысоких построек. Соответственно, думаю законом больших чисел объясянется зависимости со: стоимостью, кол-вом комнат и общей площадью.
- Действует общие тренды: чем больше площадь — тем выше ценник. Также диаграмма Количество комнат свидетельствует о том, что больше всего 3-4х комнатных квартир с общей площадью до 150 кв.м.
